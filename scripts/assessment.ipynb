```python
from models_comparison import getMtuberculosis_models
from models_comparison import ModelsComparisonMetrics

Mtuberculosis_models = getMtuberculosis_models()
modelsComparisonMetrics = ModelsComparisonMetrics(Mtuberculosis_models)
```

```python
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score, matthews_corrcoef
from assessment import ResultsReport

metrics = [balanced_accuracy_score, precision_score, recall_score, matthews_corrcoef]
reference_model = Mtuberculosis_models["reference model"]
del Mtuberculosis_models["reference model"]
report = ResultsReport(reference_model, list(Mtuberculosis_models.values()))
report.generate_report_csv(metrics, "metrics/Mtuberculosis_models_results.csv")
report.generate_radar_graph(metrics, ["Balanced Accuracy", "Precision", "Recall", "MCC"], "metrics/Mtuberculosis_radar.png")
```

```python
report.generate_confusion_matrices("metrics/Mtuberculosis_confusion_matrices")
```
